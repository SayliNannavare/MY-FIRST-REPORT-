**Marker|line|color**

**':' dotted line**

```python
import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2,13,55,4])

plt.plot(ypoints, 'o:r')
plt.show()
```

**'-' :solid line**

```python
import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2,13,55,4])

plt.plot(ypoints, 'o-r')
plt.show()
```

**'--' double dash**

```python
import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2,13,55,4])

plt.plot(ypoints, 'o--r')
plt.show()
```

**change size to maeker**

```python
import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2,13,55,4])

plt.plot(ypoints, marker='o', ms=20)
plt.show()
```

**change the maker colour **

```python
import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2,13,55,4])

plt.plot(ypoints, marker='o', ms=20 ,mfc='r')
plt.show()
```
